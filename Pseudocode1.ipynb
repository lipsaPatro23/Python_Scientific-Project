```
#Code a Sierpinski triangle
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
N = 10000

#initialize x and y

x = np.zeros(N)
y = np.zeros(N)

for i in range(1,N):
  #generate random number from the set 1,2,3 or {1,2,3}
  k = np.random.randint(1,4)

  #update x & y points
  x[i]= x[i-1]/2 + k-1
  y[i]= y[i-1]/2
  if k==2:
    y[i]+=2
plt.plot(x,y,'k.',markersize = 1)
plt.title('Sierpinski triangle')
plt.axis('off')
plt.show()


```
